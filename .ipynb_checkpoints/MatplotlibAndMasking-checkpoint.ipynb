````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
````

### Hozzuk létre a, b és c vektorokat, a végén majd ezek lesznek a 3d-ben kijelzett, generált "mérsek"

````python
# a generált "mérsek" darabszáma
numOfMeas = 200

# variancia 2.5^2 átlag 5 - 200 db
a = np.random.normal(5, 2.5, numOfMeas)

# 0 - 19.9-ig - 200 db
b = np.arange(0, numOfMeas) / 10
#print(aMask)

# színuszos
c = np.sin(np.arange(0, numOfMeas)) * 20

````

### Kijelzés 1

````python
fig1, axes = plt.subplots(1, 3, figsize=(12,3))
plt.subplot(131); plt.plot(a); plt.title("a vektor")
plt.subplot(132); plt.plot(b); plt.title("b vektor")
plt.subplot(133); plt.plot(c); plt.title("c vektor")
plt.show()
````

### Maszkolások

````python
# készítsünk randum kiugró zajt 22-es értékkel, mondjuk 5 darabot
noiseMask = np.random.randint(numOfMeas, size = 5)
for n in noiseMask:
    a[n] = 22

# maszkolás példa
aMask1 = np.ma.masked_where(a < 5, a)
aMask2 = np.ma.masked_where(a != 22, a)
aMask3 = np.ma.masked_where(a == 22, a)
bMask1 = np.ma.masked_where(b <= 100, a)
````

### Kijelzés 2

Jelezzük ki a maszkolt értékeket

````python
''' '''
fig2 = plt.figure(figsize=(10,4))
plt.plot(a, 'k', label = "Variancia 2.5^2 átlag 5 (a)")
plt.plot(aMask1, 'r*', label = "Maszkolt részek < 5 (aMask1)")
plt.plot(aMask2, 'b^', label = "Maszkolt részek != 22 (aMask2)")
#plt.plot(c, label = "C")
#plt.plot(c, label = "C")
plt.legend()
plt.show()

````

### Kijelzés 3

A randum kiugró zaj 22-es értékkel, külön láthatóvá válik hisztogramon is

````python
fig3 = plt.figure(figsize=(10,4))
plt.hist(a, bins = 50, normed = 1); plt.show()
plt.show()
````

### Kijelzés 4

3d-ben is szemlétethető a maszkolt mesterséges zaj

````python
fig4 = plt.figure("3d", figsize=(12,6))
ax1 = fig4.add_subplot(121, projection='3d')
ax2 = fig4.add_subplot(122, projection='3d')

ax1.scatter(a, b, c, label = "Generált értékek")
ax1.scatter(aMask2, b, c, label = "Zaj", marker = '^', s = 200)
ax1.legend()
ax1.set_xlabel('X Label'); ax1.set_ylabel('Y Label'); ax1.set_zlabel('Z Label')

ax2.scatter(aMask3, b, c)
ax2.set_xlabel('X Label'); ax2.set_ylabel('Y Label'); ax2.set_zlabel('Z Label')
plt.show()
````
