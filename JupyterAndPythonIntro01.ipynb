````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

x = np.arange(0, 2 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

print("Ezek az értékek: \n")
print(x, y_sin, y_cos)
````

#### A numpy-ban a `shape` a tömb méretét adja vissza. 

````python
print("x mérete: %d" % x.shape)
````

#### Így létre tudunk hozni egy ugyanakkora vektort, mint az x.

````python
y_rand = np.random.random(x.shape)
print("Ezek a random értékek: \n")
print(y_rand)
````

#### Ezt már ki tudjuk plotolni.

````python
plt.subplot(2, 1, 1)
plt.plot(x, y_sin)
plt.title('Színus')
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Koszinusz + Random')
plt.plot(x, y_rand)
plt.show()
````




#### Ennél persze sokkal több lehetőségünk van....

````python
xx = np.linspace(-0.75, 1., 100)

n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1, 4, figsize=(12,3))

axes[0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))
axes[0].set_title("scatter")

axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title("bar")

axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5);
axes[3].set_title("fill_between");
````
