Ciklusok:

````python
print("range(6):       ", end="")
# 0 1 2 3 4 5
for x in range(6):
    print("%4d" % x, end="")
    
print("\nrange(3, 9):    ", end="")
# 3 4 5 6 7 8
for x in range(3, 9):
    print("%4d" % x, end="") 

print("\nrange(3, 14, 2):", end="")
# 3 5 7 9 11 13
for x in range(3, 14, 2):
    print("%4d" % x, end="")

print("\narray:          ", end="")
primes = [2, 3, 5, 7]
for prime in primes:
    print("%4d"  % prime, end="")


print("\nwhile:          ", end="")
# 0  1  2  3  4 
i = 0
while i < 5:
    print("%4d" % i, end="")
    i += 1 
````

### Matplotlib, numpy

Ha csak pythonban használjuk akkor nem kell, de hogy a notebookban jelenjen meg kel a

````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

x = np.arange(0, 2 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

print("Ezek az értékek: \n")
print(x, y_sin, y_cos)
````

#### A numpy-ban a `shape` a tömb méretét adja vissza. 

````python
print("x mérete: %d" % x.shape)
````

#### Így létre tudunk hozni egy ugyanakkora vektort, mint az x.

````python
y_rand = np.random.random(x.shape)
print("Ezek a random értékek: \n")
print(y_rand)
````

#### Ezt már ki tudjuk plotolni.

````python
plt.subplot(2, 1, 1)
plt.plot(x, y_sin)
plt.title('Színus')
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Koszinusz + Random')
plt.plot(x, y_rand)
plt.show()
````




#### Ennél persze sokkal több lehetőségünk van....

````python
xx = np.linspace(-0.75, 1., 100)

n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1, 4, figsize=(12,3))

axes[0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))
axes[0].set_title("scatter")

axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title("bar")

axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5);
axes[3].set_title("fill_between");
````

## OpenCV

````python
import cv2

img1 = cv2.imread("dog.jpg")
plt.imshow(img1, cmap = 'gray', interpolation = 'bicubic')
# plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()

````


Valamilyen oknál fogva fordítva olvasta be RGB helyett BGR, viszont ez gyorsan javítható, a []- között a kettőspontok azt jelölik, hogy az adott dimenzióból mindent ki szeretnénk választani:

````python
img2 = img1
img2 = img1[:,:,::-1]

plt.imshow(img2, cmap = 'gray', interpolation = 'bicubic')
plt.show()
````


A MATLAB-ból ismert : szintaktika máshol is alkalmazható:

````python
img3 = img2[50:200,150:400,:]

fig2, axes2 = plt.subplots(1, 2, figsize=(12,3))

axes2[0].imshow(img3, cmap = 'gray', interpolation = 'bicubic')
axes2[1].imshow(img3)

plt.show()
````


Számos hasznos parancs van numpy-ban pl  a flipud vagy a fliplr (up-down, left-right)

````python
img4 = np.flipud(img3)
plt.imshow(img4, cmap = 'gray', interpolation = 'bicubic')
plt.show()
````
